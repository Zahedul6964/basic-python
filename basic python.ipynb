```python
#inherientence

# Parent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

# Child class
class Student(Person):
    def __init__(self, name, age, major):
        super().__init__(name, age)  # call parent constructor
        self.major = major

    def show_major(self):
        print("Major:", self.major)

# Create object
s1 = Student("Zahedul", 20, "Computer Science")
s1.display()      # parent method
s1.show_major()   # child method
```

```python
# Parent class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_info(self):
        print("Brand:", self.brand)
        print("Model:", self.model)

# Child class
class Car(Vehicle):
    def __init__(self, brand, model, doors, color):
        super().__init__(brand, model)  # call parent constructor
        self.doors = doors
        self.color=color

    def show_doors(self):
        print("Number of doors:", self.doors)
        print("Color is:", self.color)

# Child class
class Bike(Vehicle):
    def __init__(self, brand, model, type_bike):
        super().__init__(brand, model)
        self.type_bike = type_bike

    def show_type(self):
        print("Bike type:", self.type_bike)

# Create objects
car1 = Car("Toyota", "Corolla", 4,'black')
bike1 = Bike("Yamaha", "R15", "Sport")

# Access parent and child methods
car1.show_info()     # parent method
car1.show_doors()    # child method

bike1.show_info()    # parent method
bike1.show_type()    # child method
```

```python
#multilevel inheritence
# Grandparent class
class Person:
    def set_person_info(self, name, age):
        self.name = name
        self.age = age

    def show_person_info(self):
        print("Name:", self.name)
        print("Age:", self.age)

# Parent class
class Student(Person):
    def set_student_info(self, name, age, major):
        super().set_person_info(name, age)  #call parent method
        self.major = major

    def show_student_info(self):
        print("Major:", self.major)
        print("Name:", self.name)

# Child class
class Exam(Student):
    def set_exam_info(self, marks):
        self.marks = marks

    def show_exam_info(self):
        print("Marks:", self.marks)

# Create object of child class
e1 = Exam()
e1.set_student_info("Tuhin", 22, "Computer Science")
e1.set_exam_info(95)

# Access all methods
e1.show_person_info()     # from Grandparent
e1.show_student_info()    # from Parent
e1.show_exam_info()       # from Child
```
